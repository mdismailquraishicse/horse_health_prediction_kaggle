```python
def null_calculator(dataset):
    null_data = dataset.isna().sum()
    null_percentage = round(dataset.isna().sum()/len(dataset)*100,2)
    null_df = pd.DataFrame(data=null_data)
    null_df.rename(columns={0:'null_count'},inplace=True)
    null_df['null_%'] = null_percentage.values
    return null_df

null_calculator(train)
```

```python
plt.rcParams['figure.figsize'] = (5, 3)
less_effect_cols = []
high_effective_cols = []
bar_colors = ['#ffcccc','cyan', 'yellow', 'orange']
age_surgery_df = [surgery_adult_df, surgery_young_df, non_surgery_adult_df, non_surgery_young_df]
color1 = ['#d4d4d4','#e4e4e4','#f4f4f4']
titles = ['Surgery = yes & age = adult', 'Surgery = yes & age = young', 'Surgery = no & age = adult', 'Surgery = no & age = young']

# Funtion for plotting pie chart subplots
def pie_vc_plot(dfs=age_surgery_df, titles = titles, color=color1, col='outcome', rc = [2,2], rng=[2,2], figsize=(12,6), dim=0):
    fig, axes = plt.subplots(nrows=rc[0], ncols=rc[1], figsize=figsize)
    k = -1
    for i in range(rng[0]):
        if dim == 1:
            dfs[i][col].value_counts().plot.pie(ax=axes[i], autopct='%0.2f%%',colors=color)
            axes[i].set_title(titles[i])
        else:
            for j in range(rng[1]):
                k = k+1
                dfs[k][col].value_counts().plot.pie(ax=axes[i][j], autopct='%0.2f%%',colors=color)
                axes[i,j].set_title(titles[k])
    plt.show()
```

```python
high_null_cols = ['rectal_exam_feces', 'abdomen']
high_effective_cols = ['id','surgery', 'age', 'rectal_temp', 'pulse', 'respiratory_rate', 'temp_of_extremities','peripheral_pulse', 'mucous_membrane',
                       'capillary_refill_time','pain', 'abdominal_distention','abdomo_appearance','surgical_lesion','nasogastric_reflux_ph',
                       'rectal_exam_feces', 'abdomen','packed_cell_volume', 'total_protein','abdomo_protein', 'lesion_1', 'lesion_2', 'lesion_3', 'outcome']
```

```python
null_cols = ['rectal_exam_feces', 'abdomen']

null_imputing_cols = ['temp_of_extremities' ,'peripheral_pulse', 'mucous_membrane', 'capillary_refill_time',
                      'pain', 'peristalsis', 'abdominal_distention','nasogastric_tube', 'nasogastric_reflux', 'abdomo_appearance']
# train['temp_of_extremities'].describe()['top']

dummy_cols = ['surgery_no','surgery_yes','age_adult','age_young',
              'temp_of_extremities_cold','temp_of_extremities_cool','temp_of_extremities_normal','temp_of_extremities_warm',
              'peripheral_pulse_absent','peripheral_pulse_increased','peripheral_pulse_normal','peripheral_pulse_reduced',
              'mucous_membrane_bright_pink','mucous_membrane_bright_red','mucous_membrane_dark_cyanotic','mucous_membrane_normal_pink',
              'mucous_membrane_pale_cyanotic','mucous_membrane_pale_pink','capillary_refill_time_less_3_sec','capillary_refill_time_more_3_sec',
              'pain_alert','pain_depressed','pain_extreme_pain','pain_mild_pain','pain_severe_pain',
              'peristalsis_absent','peristalsis_distend_small','peristalsis_hypermotile','peristalsis_hypomotile','peristalsis_normal',
              'abdominal_distention_moderate','abdominal_distention_none','abdominal_distention_severe','abdominal_distention_slight',
              'nasogastric_tube_none','nasogastric_tube_significant','nasogastric_tube_slight','nasogastric_reflux_less_1_liter',
              'nasogastric_reflux_more_1_liter','nasogastric_reflux_none','nasogastric_reflux_slight',
              'abdomo_appearance_clear','abdomo_appearance_cloudy','abdomo_appearance_serosanguious','surgical_lesion_no','surgical_lesion_yes',
             'cp_data_no', 'cp_data_yes', 'capillary_refill_time_3', 'pain_slight']

dummy_encoding_cols = ['surgery','age','temp_of_extremities','peripheral_pulse','mucous_membrane', 'capillary_refill_time',
                       'pain','peristalsis', 'abdominal_distention','nasogastric_tube','nasogastric_reflux' ,
                       'abdomo_appearance','surgical_lesion','cp_data']

col_to_scale = ['rectal_temp','pulse', 'respiratory_rate','packed_cell_volume','total_protein','nasogastric_reflux_ph',
                'abdomo_protein','lesion_1', 'lesion_2', 'lesion_3']

def encode_target(col):
    if col=='lived':
        return 1
    elif col == 'died':
        return 0
    else:
        return 2

def outlier_remover(dataset, col):
    q1=dataset[col].quantile(.25)
    q3=dataset[col].quantile(.75)
    iqr = q3-q1
    upper = q3+(1.5*iqr)
    lower = q1-(1.5*iqr)
    dataset = dataset[(dataset[col]<upper) & (dataset[col]>lower)]
    return dataset

def scaling(dataset, cols):
    minmax_scaler = MinMaxScaler()
    result = minmax_scaler.fit_transform(dataset[cols])
    i = -1
    for col in cols:
        i = i+1
        dataset[col]=result[:,i]
    return dataset

def treat_missing(dataset, cols1, cols2):
    dataset.drop(columns= cols1, inplace=True)
    # dataset.dropna(inplace= True)
    for col in cols2:
        if col not in list(dataset.columns):
            pass
        else:
            dataset[col] = dataset[col].describe()['top']
    return dataset
    
def dummy_encoding(dataset, cols):
    dataset = pd.get_dummies(dataset, columns= cols)
    for col in dummy_cols:
        if col not in list(dataset.columns):
            pass
        else :
            dataset[col]=dataset[col].apply(lambda x:1 if x==True else 0)
    return dataset
# def preprocessing(dataset):
#     dataset = scaling(dataset, col_to_scale)
#     dataset = dataset[dataset['capillary_refill_time']!='3']
#     return dataset
```

```python
train_df = train.copy()
test_df = test.copy()
train_df = treat_missing(train_df, null_cols, null_imputing_cols)
test_df = treat_missing(test, null_cols, null_imputing_cols)
# print('train_df after removing missing', train_df.shape)
# train_df = outlier_remover(train_df, 'total_protein')
# print('train_df after removing outliers 1', train_df.shape)
# train_df = outlier_remover(train_df, 'lesion_1')
# print('train_df after removing outliers 2', train_df.shape)
X = train_df.drop(columns=['outcome'])
y = train_df['outcome']
print('len of X',X.shape)
X = scaling(X,col_to_scale)
print('len of X after scaling',X.shape)
test_df = scaling(test_df,col_to_scale)
X = dummy_encoding(X, dummy_encoding_cols)
test_df = dummy_encoding(test_df, dummy_encoding_cols)
print('len of X after encoding',X.shape)
y = y.apply(encode_target)
```
